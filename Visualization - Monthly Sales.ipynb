```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("month_sales.csv")
```

```python
# plotting 4 subplots
fig, (x1,x2,x3,x4) = plt.subplots(4,1, figsize = (7,11))

# plotting full graph in all 4 subplots
axes = [x1, x2, x3, x4]
for i in axes:
    i.plot(sales["Months"], sales["sales"], color = "#cbcff2", alpha = 0.5)
    i.set_xticklabels([])
    i.set_yticklabels([])
    
    i.tick_params(bottom = False, left = False)
    
    i.spines[["bottom", "top", "left", "right"]].set_visible(False)

# plotting quarterly data in different subplots
x1.plot(sales.iloc[:3]["Months"], sales.iloc[:3]["sales"], color = "#0212a3", linewidth = 2)
x2.plot(sales.iloc[2:6]["Months"], sales.iloc[2:6]["sales"], color = "#0212a3", linewidth = 2)
x3.plot(sales.iloc[5:9]["Months"], sales.iloc[5:9]["sales"], color = "#0212a3", linewidth = 2)
x4.plot(sales.iloc[8:12]["Months"], sales.iloc[8:12]["sales"], color = "#0212a3", linewidth = 2)

# adding title and sales in subplot 1 (1st Quarter)
x1.text(0.6, 17000, "14236.9", size = 6.5, weight = 2.5, color = "#0212a3")
x1.text(2.5, 60000, "55691.01", size = 6.5, weight = 2.5, color = "#081270")
x1.text(2.4, 13000, "Jan-Mar", size = 8, weight = 3, rotation = 63, color = "#050e63")
x1.text(5.5,6000,f"Total Sales in Q1 = {sales.iloc[:3].sum()[1]}", color = "#050e63")

# adding title and sales in subplot 2 (2nd Quarter)
x2.text(2.5, 60000, "55691.01", size = 6.5, weight = 2.5, color = "#0212a3")
x2.text(5.4, 38000, "34595.13", size = 6.6, weight = 2.5, rotation = 2, color = "#0212a3")
x2.text(4, 14000, "Apr-Jun", size = 8, weight = 3, rotation = -10, color = "#050e63")
x2.text(5.5,6000,f"Total Sales in Q2 = {round(sales.iloc[3:6].sum()[1], 2)}", color = "#050e63")

# adding title and sales in subplot 3 (3rd Quarter)
x3.text(5.4, 38000, "34595.13", size = 6.6, weight = 2.5, rotation = 2, color = "#0212a3")
x3.text(8.6, 85000, "81777.35", size = 6.5, weight = 2.5, color = "#0212a3")
x3.text(7.9, 50000, "Jul-Sep", size = 8, weight = 3, rotation = 63, color = "#050e63")
x3.text(5.5,6000,f"Total Sales in Q3 = {round(sales.iloc[6:9].sum()[1], 2)}", color = "#050e63")

# adding title and sales in subplot 4 (4th Quarter)
x4.text(8.6, 85000, "81777.35", size = 6.5, weight = 2.5, color = "#0212a3")
x4.text(11.6, 61000, "69545.62", size = 6.5, weight = 2.5, color = "#0212a3")
x4.text(10.2, 60000, "Oct-Dec", size = 8, weight = 3, rotation = 62, color = "#050e63")
x4.text(5.5,6000,f"Total Sales in Q4 = {round(sales.iloc[9:12].sum()[1], 2)}", color = "#050e63")

# adding main title and sub-title
x1.set_title("Quarterly Performance Review of Market", loc = "left", size = 13, weight = "bold")
x1.text(0.5,80000,"Total Sales in each Quarter", size = 9)
```
